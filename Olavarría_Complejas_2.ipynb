```
# First networkx library is imported 
# along with matplotlib 
import networkx as nx 
import matplotlib.pyplot as plt 


# Defining a Class 
class GraphVisualization: 

	def __init__(self): 
		
		# visual is a list which stores all 
		# the set of edges that constitutes a 
		# graph 
		self.visual = [] 
		
	# addEdge function inputs the vertices of an 
	# edge and appends it to the visual list 
	def addEdge(self, a, b): 
		temp = [a, b] 
		self.visual.append(temp) 
		
	# In visualize function G is an object of 
	# class Graph given by networkx G.add_edges_from(visual) 
	# creates a graph with a given list 
	# nx.draw_networkx(G) - plots the graph 
	# plt.show() - displays the graph 
	def visualize(self): 
		G = nx.Graph() 
		G.add_edges_from(self.visual) 
		nx.draw_networkx(G) 
		plt.show() 

# Driver code 
G = GraphVisualization() 


for i in range(len(A)):
  for j in range(len(A[i])):
    if A[i][j] == 1:
      G.addEdge(i, j) 

#Visualización de red:
G.visualize() 
```

### **Pregunta 5**

Para que $P(k) = C \alpha^k$ sea una función de distribución tiene que estar normalizada:

$$\sum_0^\inf C \alpha^k = 1$$
$$ \frac{C}{1-\alpha^k} = 1$$
$$ C= 1-\alpha^k$$

Su función generadora sería:

$$\hat{P}(k) = \sum_0^k (1-\alpha)\alpha^k$$
$$           = \sum_0^k \alpha-\alpha^{k+1}$$
$$           =  \alpha^0 - \alpha^1 + \alpha^1 - ... -\alpha^k + \alpha^k -\alpha^{k+1}$$
$$           =  1 -\alpha^{k+1}$$


Para que exista una componente conexa el valor esperado de los grados debiese ser mayor a 1:

$$|E(x)| = (1-\alpha)\sum_0^\inf k\alpha^k$$
$$ = (1-\alpha)\frac{\alpha}{(1-\alpha)^2}$$
$$ 1<\frac{\alpha}{1-\alpha}$$
$$ \frac{1}{2} < \alpha$$

Entonces si $\alpha > 0.5$ se tiene una componente conexa, en caso contrario no se tendría una (estadísticamente).

```
import igraph
import random
random.seed(765)

def master_fun(g,red="dolphins.gml",attack="random"):
  print("Red",red, "atacada con:", attack)
  initial_len = len(g.components()[0])
  print("Largo de cluster inicial: ", initial_len)
  count = 0
  if attack == "random":
      while len(g.components()[0]) > initial_len/2:
        count = count + 1
        i = random.randint(0, len(g.components()[0])-1)
        g.delete_vertices([i])
  elif attack == "degree":
    while len(g.components()[0]) > initial_len/2:
      count = count + 1
      degrees = g.degree()
      g.delete_vertices([degrees.index(max(degrees))])

  elif attack == "betweenness":
    while len(g.components()[0]) > initial_len/2:
      count = count + 1
      betweenness = g.betweenness()
      g.delete_vertices([betweenness.index(max(betweenness))])
      

  print("Porcentaje necesitado:", str(round(100*count/initial_len,1))+"%")
  print("")
Dol = igraph.Graph.Read_GML("dolphins.gml")
Gnu = igraph.Graph.Read_GML("gnutella.gml")
Er1 = igraph.Graph.Erdos_Renyi(n=100,p=0.7)
Er2 = igraph.Graph.Erdos_Renyi(n=100,p=0.9)
```

```
master_fun(Dol, "Dolphins", "random")
master_fun(Dol, "Dolphins", "degree")
master_fun(Dol, "Dolphins", "betweenness")
```

```
master_fun(Gnu, "Gnutella", "random")
master_fun(Gnu, "Gnutella", "degree")
master_fun(Gnu, "Gnutella", "betweenness")
```

```
master_fun(Er1, "Erdos Renyi 70%", "random")
master_fun(Er1, "Erdos Renyi 70%", "degree")
master_fun(Er1, "Erdos Renyi 70%", "betweenness")
```

```
master_fun(Er2, "Erdos Renyi 90%", "random")
master_fun(Er2, "Erdos Renyi 90%", "degree")
master_fun(Er2, "Erdos Renyi 90%", "betweenness")
```

De los resultados se puede ver que para redes estrucuradas es más eficiente un ataque priorizando betweenness > grado del nodo > aleatorio. Esto hace centido ya que el nivel de betweenness indica el nodo más central de la red, el cual tiende a ser el más conectado e importante. En la red de gnutella se desarma la red con menos del 1% de los nodos. En el peor de los casos se necesita el 50% para que esto pase en promedio. Este comportamiento se nota en las redes generadas aleatoriamente, ya que carecen de estructura.
